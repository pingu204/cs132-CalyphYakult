```python
# @title Libraries
import numpy as np
import pandas as pd
import geopandas as gpd

import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import seaborn as sns

import datetime
```

SETUP

```python
PHL = gpd.read_file("./Philippines_AL258.kml", driver='KML')
PHL.rename(columns={"Name":"region_name"}, inplace=True)

df_pop = pd.read_csv('./regional_budget.csv')
```

DTI (2020-2024)

```python
df_pop['dti_result'] = df_pop['DTI']/df_pop['GRDP']
```

DOTr (2020-2024)

```python
df_pop['dotr_result'] = df_pop['DOTr']/df_pop['Population']
```

```python
plt.style.use("dark_background")

df_pop_2020 = (df_pop.loc[df_pop["YEAR"] == 2020]).drop(columns=["YEAR"])
gdf = pd.merge(PHL, df_pop_2020, on=['region_name'])

fig, ax = plt.subplots(figsize=(16, 8))

# Plot choropleth map
vmin =  min(df_pop_2020["dotr_result"].tolist())
vmax = max(df_pop_2020["dotr_result"].tolist())
g = gdf.plot(column='dotr_result', cmap='flare_r', vmin=vmin, vmax=vmax, ax=ax)

# Clean up
g.set_axis_off()
g.spines[['left', 'bottom', 'right', 'top', ]].set_visible(False)

# Add colorbar
# https://stackoverflow.com/a/36080553
f = g.get_figure()
cax = f.add_axes([0.65, 0.15, 0.03, 0.61])
sm = plt.cm.ScalarMappable(cmap='flare_r', norm=plt.Normalize(vmin=vmin, vmax=vmax))
# sm._A = [] # what for?
cbar = f.colorbar(sm, cax=cax)
# cbar.set_label('Inequality Index', color)
cbar.outline.set_visible(False)
f.set_facecolor('#1b181c')
# Add titles, labels, and legends
g.set_title('DOTr Budget Pesos per person 2020',color="white")
plt.show()
```
