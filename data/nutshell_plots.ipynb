```python
# @title Libraries
import numpy as np
import pandas as pd
import geopandas as gpd

import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import seaborn as sns

import datetime
```

```python
# @title Displaying geospatial data
plt.style.use("dark_background")
# Load PH map
PHL = gpd.read_file("./Philippines_AL258.kml", driver='KML')
PHL.rename(columns={"Name":"Region_name"}, inplace=True)
df_pop = pd.read_csv('./ineq_index.csv')

# Merge dataframes
gdf = pd.merge(PHL, df_pop, on=['Region_name'])
# # gdf['population_2015m'] = gdf['population_2015']/1.e6 # convert to million unit

fig, ax = plt.subplots(figsize=(16, 8))

# Plot choropleth map
vmin, vmax = 0,1
g = gdf.plot(column='range', cmap='flare_r', vmin=vmin, vmax=vmax, ax=ax)

# Clean up
g.set_axis_off()
g.spines[['left', 'bottom', 'right', 'top', ]].set_visible(False)

# Add colorbar
# https://stackoverflow.com/a/36080553
f = g.get_figure()
cax = f.add_axes([0.65, 0.15, 0.03, 0.61])
sm = plt.cm.ScalarMappable(cmap='flare_r', norm=plt.Normalize(vmin=vmin, vmax=vmax))
# sm._A = [] # what for?
cbar = f.colorbar(sm, cax=cax)
# cbar.set_label('Inequality Index', color)
cbar.outline.set_visible(False)
f.set_facecolor('#1b181c')
# Add titles, labels, and legends
g.set_title('2024',color="white")

plt.show()
```
